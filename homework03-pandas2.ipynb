### Задание 1.2
- Создать новый признак delta_renovated, который будет содержать разницу в годах между годом реновацией дома и годом постройки дома.
- Если реновации дома не было, то в новом признаке поставьте 0

```python
diff_renovated = []
for i, row in df.iterrows():
    diff = 0 if row['yr_renovated'] == 0 else row['yr_renovated'] - row['yr_built']
    diff_renovated.append(diff)
diff_renovated[:5]
```

```python
df['delta_renovated'] = diff_renovated
```

```python
df[df['delta_renovated'] > 0].head()
```

### Задание 1.3
- Создайте признаки года продажи, месяца продажи

```python
monthes = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
df['sale_year'] = df['date'].apply(lambda my_date: my_date[:4])
df['sale_month'] = df['date'].apply(lambda my_date: monthes[int(my_date[4:6]) - 1])
```

```python
df.sample(10)
```

### Задание 3.1
- Найдите среднюю стоимость домов в зависимости от количества спален. Отсортируйте от меньшей стоимости к большей

```python
df.groupby('bedrooms').agg({'price': 'mean'}).sort_values('price', ascending=False)
```

### Задание 3.2
- Найдите минимальную, среднюю и максимальную стоимости домов в зависимости от состояния дома

```python
df.groupby('condition').agg({'price': ['min', 'mean', 'max']}).sort_values(('price', 'mean'))
```

### Задание 3.3
- Постройте таблицу с подсчетом количества домов в данных в зависимости от вида на набережную и оценкой вида

```python
df.pivot_table(index='waterfront', columns='view', aggfunc='count', values='id')
```

### Задание 3.4
- Каких домов в зависимости от этажности и количества спален больше?

```python
pd.crosstab(index=df['floors'], columns=df['bedrooms']).max()
```

### Задание 3.5
